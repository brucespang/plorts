```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
%config InlineBackend.figure_format = 'retina'
import plorts
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
plt.style.use(['plorts', 'plorts-web'])

xmax = 2*np.pi
f = lambda x,offset: (offset+1)*np.sin(x-offset)
xs = np.arange(0,xmax,0.1)
offsets = range(0,len(plorts.palettes.solarized))
offsets = [4,5,6,7]
df = pd.DataFrame([[x, f(x, offset), i, np.abs(f(x, offset))] for x in xs for i,offset in enumerate(offsets)], columns=["x", "y", "offset", "abs"])
```

```python
plt.figure()
plt.title("Line Plot Example")
plorts.plot(df, x="x", y="y", hue="offset")
plorts.legend(loc='end')
plt.ylabel("Y label")
plt.xlabel("X label")
plorts.style_axis()
plorts.savefig("output/line-plot.png")
```

```python
plt.figure()
plt.title("Scatter Plot Example")
plorts.scatter(df, x="x", y="y", hue="offset")
plorts.legend(loc='end')
plt.ylabel("Y label")
plt.xlabel("X label")
plorts.style_axis()
plorts.savefig("output/scatter-plot.png")
```

```python
plt.figure()
sns.boxplot(data=df, x="offset", y="y")
plt.ylabel("Y label")
plt.xlabel("X label")
plt.title("Boxplot Example")
plorts.style_axis()
plorts.savefig("output/boxplot.png")
```

```python
plt.figure()
plt.hist(df.y, rwidth=0.98, color='C4')
plt.ylabel("Y label")
plt.xlabel("X label")
plt.title("Histogram Example")
plorts.style_axis()
plorts.savefig("output/histogram.png")
```

```python
plorts.stackplot(df, x="x", y="abs", hue="offset")
plt.legend(loc='best')
plt.ylabel("$|\\sin(x)|$")
plt.title("Stacked Plot Example")
plorts.style_axis()
plorts.savefig("output/stack-plot.png")
```

```python
plt.figure()
x = np.arange(-1,4,0.001)
y = np.minimum(-1*x+2, 1)
bottom = np.minimum(y,0)
plt.plot(x,y,color="C3")
plt.plot([0,0],[-1,3],color="C3")
plt.plot([-1,3],[0,0],color="C3")
plt.axvline(x=1.5, color="C3")
plt.fill_between(x,bottom,y, where=(x>=0)&(x<=1.5),color="C1")
plt.axis(xmin=-0.1, xmax=2.5, ymin=-0.3, ymax=2.3)
plt.title("Feasible Region")
plt.xlabel("x")
plt.ylabel("y")
plorts.style_axis(show_xaxis=True)
```
