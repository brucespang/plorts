```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
%config InlineBackend.figure_format = 'retina'
import plorts
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
xmax = 2*np.pi
f = lambda x,offset: (offset+1)*np.sin(x-offset)
xs = np.arange(0,xmax,0.1)
offsets = range(0,len(plorts.palettes.solarized))
offsets = [4,5,6,7]
df = pd.DataFrame([[x, f(x, offset), i] for x in xs for i,offset in enumerate(offsets)], columns=["x", "y", "offset"])
```

```python
import plorts
reload(plorts)
plt.style.use('./plorts/plorts.mplstyle')
plt.style.use('./plorts/plorts-web.mplstyle')
plt.figure()
plt.title("Some Nice Sine Curves")
plorts.plot(df, x="x", y="y", hue="offset")
plorts.legend(loc='end')
plt.ylabel("Y label")
plt.xlabel("X label")
plt.axis(ymin=-8.5,ymax=8.5)
plorts.style_axis()
```

```python
plt.style.use('./plorts/plorts.mplstyle')
plt.style.use('./plorts/plorts-web.mplstyle')
plt.figure()
plt.title("Some Nice Sine Curves", weight="bold", y=1.1, x=-0.04, ha='left')
plorts.scatter(df, x="x", y="y", hue="offset")
plorts.legend(loc='end')
plt.ylabel("Y label")
plt.xlabel("X label")
plt.axis(ymin=-8.5,ymax=8.5)
plorts.style_axis()
```

```python
plt.figure()
sns.boxplot(data=df, x="offset", y="y")
plt.ylabel("Y label")
plt.xlabel("X label")
plt.title("Some Nice Sine Curves")
plorts.style_axis()
```

```python
import plorts
df['abs'] = np.abs(df.y)
plorts.stackplot(df, x="x", y="abs", hue="offset")

plt.legend(loc='best')
plt.ylabel("$|\\sin(x)|$")
plorts.style_axis()
```

```python
plt.figure()
x = np.arange(-1,4,0.001)
y = np.minimum(-1*x+2, 1)
bottom = np.minimum(y,0)
plt.plot(x,y,color="C3")
plt.plot([0,0],[-1,3],color="C3")
plt.plot([-1,3],[0,0],color="C3")
plt.axvline(x=1.5, color="C3")
plt.fill_between(x,bottom,y, where=(x>=0)&(x<=1.5),color="C1")
plt.axis(xmin=-0.1, xmax=2.5, ymin=-0.3, ymax=2.3)
plt.title("Feasible Region")
plt.xlabel("x")
plt.ylabel("y")
plorts.style_axis(show_xaxis=True)
```
